## 3. Calculating the Number of Daylight Hours (10 points)

There are a few different models, with varying degrees of accuracy, to estimate the number of daylight hours at a given time and place for application in agriculture, ecology, and earth system modeling. 

A relatively simplistic one from Brock (1981) is as follows:

First, the declination of the sun is given by:

$\delta = .40928 \sin{(2\pi \frac{283+J}{365})}$

where $J$ is the day of the year (number of days since January 1st). The sunset/sunrise hour angle is then computed as 

$\beta = \arccos(-\tan{\phi}\tan{\delta})$

Where $\phi$ is the latitude in radians. Finally the daylength, in hours, is calculated

$D = 24\frac{\beta}{\pi}$

**3.1** Create a function named `daylength` that computes number of daylight hours at a given time and place. Your input parameters should be `lat` and `day` (for day of the year). Keep in mind that latitude is usually given in degrees but the equation requires latitude in radians.

```python
import math 

def daylength(lat, day):
    lat_rd = lat * (math.pi / 180)
    declination = 0.40928 * math.sin(2* math.pi * (283 + day)/365)
    angle = math.acos(- math.tan(lat_rd) * math.tan(declination))
    daylen = 24 * angle/math.pi
    return daylen
```

## 5. ArcAscii to NetCDF4 (20 points)

The directory `/data/MSWEP/AAIGrid` contains spatial grids of the [MSWEP](http://www.gloh2o.org/) dataset at a 1 month temporal resolution and 1 arcdegree spatial resolution for 2015 in the AAIGrid format. Create a single NetCDF4 file with one 3D array (time,lat,lon) that contains this data. Ensure that the created NetCDF4 is CF-compliant by following the guidelines described [here](http://cfconventions.org/Data/cf-conventions/cf-conventions-1.7/cf-conventions.html#_netcdf_files_and_components). You can also use the ERA-Interim NetCDF4 file as an example (/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc). Note that you will have to write your own function to parse the ArcAscii (i.e., AAIGRID) files; you may **not** use rasterio for this problem.

```python
import numpy as np
import glob

files = glob.glob("/data/MSWEP/AAIGrid/*")
datas = []
time = []

def read_aaigrid(file_path):
    # Initialize variables
    ncols = nrows = xllcorner = yllcorner = cellsize = nodata_value = None
    
    with open(file_path, 'r') as file:
        # Read the header
        for _ in range(6):
            line = file.readline().strip()
            key, value = line.split()
            if key.lower() == 'ncols':
                ncols = int(value)
            elif key.lower() == 'nrows':
                nrows = int(value)
            elif key.lower() == 'xllcorner':
                xllcorner = float(value)
            elif key.lower() == 'yllcorner':
                yllcorner = float(value)
            elif key.lower() == 'cellsize':
                cellsize = float(value)
            elif key.lower() == 'nodata_value':
                nodata_value = float(value)
        
        # Read the data into a NumPy array
        data = np.loadtxt(file, dtype=float)
        
    metadata = {
        'ncols': ncols,
        'nrows': nrows,
        'xllcorner': xllcorner,
        'yllcorner': yllcorner,
        'cellsize': cellsize,
        'nodata_value': nodata_value
    }
    return data, metadata

for file in files:
    tl = file.split('/')[-1].split('.')[0]
    tl_time = datetime.datetime.strptime(tl,'%Y_%m')
    time.append(tl_time)
    data, metadata = read_aaigrid(file)
    datas.append(data)

time = np.array(time)
d = np.array(datas)
```

```python
import netCDF4 as nc
import os

f = 'MSWEP_grid.nc'
os.system('rm -f %s' % f)
ff = nc.Dataset(f,'w')


ff.createDimension('time', size = d.shape[0])
ff.createDimension('lat', size = d.shape[1])
ff.createDimension('lon', size = d.shape[2])
```

```python
ff.createVariable('mon_prep','f8',dimensions=('time','lat','lon'))
ff['mon_prep'][:] = d[:]
ff['mon_prep'].units = 'mm'
ff['mon_prep'].missing_value = -9999
ff['mon_prep'].standard_name = 'monthly precipitation'
ff['mon_prep'].long_name = 'monthly precipitation'
```

```python
ff.createVariable('lon', 'i4', dimensions=('lon',))
ff['lon'][:] = np.arange(0,360)
ff['lon'].units = 'degrees_east'
ff['lon'].standard_name = 'longitude'
ff['lon'].long_name = 'longitude'
ff['lon'].axis = 'X'
```

## 6. Advanced Matplotlib: Showing Raster Data and Point Data for Temperature in Colorado (20 points)

Often in geospatial analysis, you may have access to high quality in-situ (point) data and lower quality satellite or reanalysis data with greater spatial coverage. Comparing the two is a common exercise. Using the reanalysis data for air temperature in August 2019 from `/data/PCF` as well as the 1 week half-hourly timeseries of tower air temperature reported from the 4 sites of the National Ecological Observation Network (NEON) in Colorado located here `/data/neon/temperature`, reverse engineer the code that created the following figure. 

Note that the `.tif` files are named according to the day, and have hourly data such that `read(1)` will provide the raster for 0:00 UTC, `read(2)` will provide the raster for 1:00 UTC etc. Also note that the NEON files have a `time` in seconds since January 1st 1970 (UTC), which will need to be converted to `datetime` objects (perhaps using `datetime.timedelta`)




```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import rasterio
import h5py
```

```python
sites = {'CPER':{'path':'CPER_TEMP.h5'},
        'RMNP':{'path':'RMNP_TEMP.h5'},
         'STER':{'path':'STER_TEMP.h5'},
         'NIWO':{'path':'NIWO_TEMP.h5'}}
dates = ['20190823', '20190825', '20190827']
```

```python
for site in sites:
    file = '/data/neon/temperature/'+ sites[site]['path']
    fp = h5py.File(file, "r")
    sites[site]['time'] = fp['TIME'][:]
    sites[site]['temp'] = fp['T_SONIC'][:]
    sites[site]['lat'] = fp.attrs['lat']
    sites[site]['lon'] = fp.attrs['lon']

    def seconds_to_date(seconds):
        return datetime.datetime(1970, 1, 1) + datetime.timedelta(seconds=seconds)

    sites[site]['date'] = np.array([seconds_to_date(sec) for sec in fp['TIME'][:]])
```

```python

fig = plt.figure(figsize=(15,8))
G = gridspec.GridSpec(2,3)
cmap = plt.get_cmap('coolwarm')


#rater plots 
min_value = 300
max_value = 200
for i in range(3):
    ax = fig.add_subplot(G[0,i])
    path = "/data/PCF/PCF_" + dates[i] +".tif"
    fp = rasterio.open(path)
    temp_img = fp.read(23)
    if temp_img.min() < min_value:
        min_value = temp_img.min()
    if temp_img.max() > max_value:
        max_value = temp_img.max()
    norm = lambda x: (x-min_value)/(max_value-min_value)
    img = ax.imshow(temp_img, cmap=cmap, vmin=min_value, vmax=max_value)
    plt.axis('off')
    plt.title(dates[i]+' 22:00 UTM', fontsize=15)
    for site in sites:
        lon, lat = sites[site]['lon'], sites[site]['lat']
        x, y = fp.index(lon, lat)
        date = datetime.datetime.strptime(dates[i],'%Y%m%d')
        date = date + datetime.timedelta(hours = 22)
        idex = np.argmin(np.abs(sites[site]['date']-date))
        point_value = sites[site]['temp'][idex,]
        plt.scatter(y,x, color = cmap(norm(point_value)), s=100, edgecolor='black', label='Points')
        plt.text(y+5,x-3,site,ha='left',va='center')


#line plot
ax4 = fig.add_subplot(G[1,:])
for site in sites:
    ax4.plot(sites[site]['date'], sites[site]['temp'], lw=2)
    for date in dates:
        date = datetime.datetime.strptime(date,'%Y%m%d')
        date = date + datetime.timedelta(hours = 22)
        idex = np.argmin(np.abs(sites[site]['date']-date))
        plt.scatter(sites[site]['date'][idex,], sites[site]['temp'][idex,], color='black', zorder=5, label='Points')

plt.xticks(fontsize=10,rotation=35)
plt.xlabel('Time',fontsize=15)
plt.ylabel('Temperature (K)',fontsize=15)
plt.legend(['CPER','RMNP','STER','NIWO'],fontsize=10, loc='upper right')

# color bar 
cax = fig.add_axes([0.92, 0.2, 0.02, 0.6])
cb = fig.colorbar(img, cax=cax, orientation='vertical')
cb.set_label('Temperature (K)', fontsize=15)
cb.ax.set_yticklabels(["{:.0f}".format(i) for i in cb.get_ticks()])
#Show the plot
plt.show()
```

```python
import netCDF4 as nc
import numpy as np
import datetime 
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

%matplotlib inline


file = '/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc'
fp = nc.Dataset(file)
ppt = fp['precip'][:]

m_name = ['Januarary', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September','October','November', 'December']

fig, ax = plt.subplots(figsize=(10,10))
img = ax.imshow(np.flipud(np.mean(ppt[::12,:,:], axis=0)), cmap=plt.get_cmap('RdBu'))
cb = fig.colorbar(img, orientation='horizontal',shrink=0.9,pad=0.01)
cb.ax.tick_params(labelsize=10)
cb.set_label('Monthly Precipitation (mm/day)',fontsize=15)

def animate(i):
    plt.axis('off')
    tmp = ppt[i-1::12,:,:]
    tmp = np.mean(tmp, axis=0)
    ax.imshow(np.flipud(tmp),cmap = plt.get_cmap('RdBu'))
    ax.set_title('ERA-Interim Monthly Precipitation 1979 - 2015 (%s)' % (m_name[i]),fontsize=20)
    return img,



animation = FuncAnimation(fig, animate, frames=12, interval=1000, blit=True)
video = animation.to_html5_video()
# video = animation.to_jshtml()
html = HTML(video)
display(html)
plt.close()
```
