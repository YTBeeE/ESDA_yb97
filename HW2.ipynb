**1.1** Prove that $E[aX + b] = aE[X] + b$

We assume $X$ is a continuous variable. 

$E[aX + b] = \int_{-\infty}^{\infty} (ax+b) f(x) \,dx = \int_{-\infty}^{\infty} ax f(x) \,dx + \int_{-\infty}^{\infty} b f(x) \,dx$

a and b are constants. Thus, 

$E[aX + b] = a\int_{-\infty}^{\infty} x f(x) \,dx + b\int_{-\infty}^{\infty} f(x) \,dx$

Because $\int_{-\infty}^{\infty} x f(x) \,dx = E[X]$ by definition, and 
$\int_{-\infty}^{\infty} f(x) \,dx$ is the sum of all probabilities of X, which equals 1:

$E[aX + b] = aE[X] + b$

Same reasoning if $X$ is a discrete variable. 

**1.2** Prove that $Var(X) = E[X^2] - E[X]^2$.

By definition:

$Var(X) = \int_{-\infty}^{\infty} f(x)(x-\mu)^2\,dx$

where $\mu$ is the population mean, which is a constant and equal to $E[X]$.

$Var(X) = \int_{-\infty}^{\infty} f(x)(x^2 - 2\mu x +\mu^2)\,dx$

Because $\mu$ is a constant:

$Var(X) = \int_{-\infty}^{\infty} f(x)x^2\,dx - 2\mu \int_{-\infty}^{\infty} f(x) x\,dx +\mu^2 \int_{-\infty}^{\infty} f(x)\,dx$

Again, $\int_{-\infty}^{\infty} x f(x) \,dx = E[X] = \mu$ by definition, and 
$\int_{-\infty}^{\infty} f(x) \,dx$ is the sum of all probabilities of X, which equals 1

$Var(X) = \int_{-\infty}^{\infty} f(x)x^2\,dx - 2\mu^2 +\mu^2 $

Because $\int_{-\infty}^{\infty} f(x)x^2\,dx = E[X^2]$,

$Var(X) = E[X^2] - \mu = E[X^2] - E[X]^2$

$E[X]$ and $E[Y]$ are the expectation, or population mean, of variable $X$ and $Y$, which are constants. Thus,

$Cov(X,Y) = E[XY] - E[Y]E[X] + E[Y]E[X] + E[X]E[Y] = E[XY] - E[X]E[Y]$

As a result,

# 3. Calculating the correct spatial mean (20 pts)

A simple spatial mean of a regular lat/long grid (Plate Carree) can lead to large discrepancies in the estimated spatial mean. To assess this impact, we are going to use the variable `t2m` in the annual ERA-Interim database (`/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc_ann`). Start by assembling an array of actual area per pixel and then compute for each year the area weighted mean. Then compare the resulting time series to that resulting from using the equally weighted spatial mean (i.e., naive approach). Provide an explanation of your results.

```python
import netCDF4 as nc
import numpy as np
path = "/data/era-interim/era_interim_monthly_197901_201512_upscaled.nc_ann"
era = nc.Dataset(path,'r')
# extract data
lat = era["lat"][:]
lon = era["lon"][:]
time = era["time"]
t2m = era["t2m"][:]
date = nc.num2date(time[:], units=time.units, calendar=time.calendar)
```

```python
# create a function to calculate surface area with given latitudes and longitudes
def surface_area(Re,phi_a,phi_b,lambda_c,lambda_d):
    #Convert to radians
    phi_a = phi_a*np.pi/180.0
    phi_b = phi_b*np.pi/180.0
    lambda_c = lambda_c*np.pi/180.0
    lambda_d = lambda_d*np.pi/180.0
    #Calculate surface area
    return Re**2*np.abs(np.sin(phi_a)-np.sin(phi_b))*np.abs(lambda_c - lambda_d)
```

```python
Re = 6371.0 #km
area = []
# calculate surface area of each pixel 
for a in range(len(lat)):
    for o in range(len(lon)):
        if a == 0:
            lat1 = -90
        else:
            lat1 = (lat[a-1]+lat[a])/2
        if o == 0:
            lon1 = 0
        else:
            lon1 = (lon[o-1]+lon[o])/2
        if a == len(lat) - 1:
            lat2 = 90
        else:
            lat2 = (lat[a]+lat[a+1])/2
        if o == len(lon) -1:
            lon2 = 360
        else:
            lon2 = (lon[o]+lon[o+1])/2
        area.append(surface_area(Re, lat1, lat2, lon1, lon2))
# reshape to the same shape of t2m dataset 
area = np.array(area).reshape((len(lat),len(lon)))
```

```python
simple_mean = []
weighted_mean = []
# calculate the mean of each timestamp 
for d in range(len(date)):
    data = t2m[d,:,:]
    smean = data.mean()
    wdata = data*area
    wmean = wdata.sum()/area.sum()
    simple_mean.append(smean)
    weighted_mean.append(wmean)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Create a figure and axis
plt.figure(figsize=(10, 6))
# Plot the equally weighted mean 
plt.plot(date, simple_mean, label='Equally Weighted Mean', marker='o')
# Plot the area weighted mean 
plt.plot(date, weighted_mean, label='Area Weighted Mean', marker='s')

# Adding labels and title
plt.xlabel('Date', fontsize=20)
plt.ylabel('Temperature (°C)', fontsize=20)
plt.title('Spatial Mean of Temperature Comparison Over Time', fontsize=15)

# Adding a legend to differentiate the datasets
plt.legend()

# Format x-axis to display year and month
plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%Y-%m'))

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

**Explanation**

As the latitude moves to the poles, the surface area of euqal latitude and longitude gets distorted and becomes smaller. Thus, the euqally weighted mean overweigh the temperature of polar areas where tend to have colder temperature, resulting a much smaller number than the area weighted mean temperature.

```python
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML

# Define the plot extent
img_extent = (-180, 180, -90, 90)

# Create the figure and projection (Orthographic for globe)
fig = plt.figure(figsize=(10, 10))
vmin = np.min(ppt)
vmax = np.max(ppt)
nframes = len(time_fl)

def animate(i):
    plt.clf()
    ax = plt.axes(projection=ccrs.Orthographic(central_longitude=lon[i], central_latitude=20))
    
    # Add topographic basemap
    ax.stock_img() 
    ax.add_feature(cfeature.LAND)
    ax.add_feature(cfeature.COASTLINE)

    # Plot precipitation data
    imin = np.argmin(np.abs(date-time_fl[i]))
    ppt1 = np.ma.masked_array(ppt[imin,:,:],ppt[imin,:,:]==0)
    im = plt.imshow(np.flipud(ppt1), vmin=vmin, vmax=vmax,
                    transform=ccrs.PlateCarree(),
                    cmap=plt.get_cmap('gist_earth'), extent=img_extent)

    # Add country borders
    ax.add_feature(cartopy.feature.BORDERS, linestyle=':')
    # Add flight track
    lon_ud = lon[:i+1]
    lat_ud = lat[:i+1]
    ax.plot(lon_ud, lat_ud, color='blue', linewidth=2, marker='o', transform=ccrs.Geodetic())

    # Add a colorbar for the precipitation map
    cb = plt.colorbar(im, ax=ax, orientation='horizontal', shrink=0.6, pad=0.03)
    cb.ax.tick_params(labelsize=20)
    cb.ax.set_xlabel('Surface Precipitation Rate [$kg/m^2/s$]', fontsize = 15)

    # Add title
    ax.set_title(f'{time_fl[i]}', fontsize=25)

    return(im,)

anim = animation.FuncAnimation(fig, animate, frames=nframes, interval=500, blit=True)
HTML(anim.to_html5_video())
```
